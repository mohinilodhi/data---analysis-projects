```python
import pandas as pd

data = {
    "name": ["Aman", "Riya", "Rahul", "Sneha", "Vikram", "Neha", "Amit", "Pooja", "Rohan", "Simran"],
    "gender": ["M","F","M","F","M","F","M","F","M","F"],
    "age": [22, 19, 35, 28, 40, 18, 50, 30, 27, 24],
    "survived": [1,1,0,1,0,1,0,1,1,0],
    "fare": [100,200,50,300,20,400,30,250,150,500]
}

df = pd.DataFrame(data)

import matplotlib.pyplot as plt

counts = df["survived"].value_counts()

plt.bar(counts.index, counts.values)
plt.show()
```

```python
import matplotlib.pyplot as plt

counts = df["survived"].value_counts()

plt.bar(counts.index, counts.values)
plt.title("titanic survival count")
plt.xlabel("survived(0 = no, 1= yes)")
plt.ylabel("number of Passenger")
plt.show()
```

```python
import pandas as pd
male_survived = df[(df["gender"] == "M") & (df["survived"] == 1)].shape[0]
print(male_survived)
female_survived = df[(df["gender"] == "F") & (df["survived"] == 1)].shape[0]
print(female_survived)
```

```python
import matplotlib.pyplot as plt

labels = ["Male", "Female"]
values = [male_survived, female_survived]

plt.bar(labels, values, color=["blue", "pink"])

plt.title("Survival by Gender")
plt.xlabel("Gender")
plt.ylabel("Survived Count")

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.hist(df["age"])
plt.title("Age Distribution of Titanic Passengers")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.scatter(df["age"], df["fare"])

plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.scatter(df["fare"], df["survived"])

plt.title("Fare vs Survival")
plt.xlabel("Fare")
plt.ylabel("Survived (0/1)")

plt.show()
```
