```python
import pandas as pd

data = {
    "Nmae": ["aakansha","amit","akash","riya","vinay","anushka","anjali"],
    "age" : [12,13,14,15,16,17,18],
    "class" : [ "12","12","11","12","12","12","12"],
    "gender" : ["female","male","male","female","male","female","female"],
    "maths score" :[90,89,78,88,68,99,69],
    "science score" : [90,89,78,88,68,99,69],
    "social score" : [90,89,78,88,68,99,69]
}

df = pd.DataFrame(data)
df.to_csv("student.csv", index=False)
df.head()
```

FILTERING THE
DATA

```python
df.isnull().sum()
df = df.dropna()
```

```python
df['maths score'] = df['maths score'].astype(int)
```

```python

df['maths score'].fillna(df['maths score'].mean())
```

```python
df['social score'].fillna(df['social score'].mean())
```

```python
df =df.drop_duplicates()
```

```python
df['gender'].str.lower().str.strip()
```

```python
df.columns = df.columns.str.strip()
```

```python
df.describe()
```

```python
df['gender'].value_counts().plot(kind='bar')
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
df.plot(x="Nmae",y="age",kind="bar")
plt.show()
```

```python
df["age"].plot(kind ="hist")
plt.show()
```

```python
import matplotlib.pyplot as plt
df['gender'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribution of Gender')
plt.ylabel('')
plt.show()
```

```python
plt.scatter(df['maths score'], df['science score'])
plt.xlabel('Maths Score')
plt.ylabel('Science Score')
plt.title('Scatter Plot of Maths Score vs Science Score')
```

```python
plt.boxplot(df['maths score'])
plt.xlabel('Maths Score')
plt.title('Box Plot of Maths Score')
```

```python
plt.fill_between(df['maths score'], df['science score'])
plt.xlabel('Maths Score')
plt.ylabel('Science Score')
plt.title('Fill Between Plot of Maths Score')
```

```python
plt.plot(df['maths score'], df['science score'])
plt.xlabel('Maths Score')
plt.ylabel('Science Score')
plt.title('Line Plot of Maths Score vs Science Score')
```
