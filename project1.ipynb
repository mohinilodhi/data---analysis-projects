```python
import pandas as pd

data = {
    "Nmae": ["aakansha","amit","akash","riya","vinay","anushka","anjali"],
    "age" : [12,13,14,15,16,17,18],
    "class" : [ "12","12","11","12","12","12","12"],
    "gender" : ["female","male","male","female","male","female","female"],
    "maths score" :[90,89,78,88,68,99,69],
    "science score" : [90,89,78,88,68,99,69],
    "social score" : [90,89,78,88,68,99,69]
}

df = pd.DataFrame(data)
df.to_csv("student.csv", index=False)
df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df.isnull().sum()
df = df.dropna()
```

```python
df['maths score'] = df['maths score'].astype(int)
```

```python

df['maths score'].fillna(df['maths score'].mean())
```

```python
df['social score'].fillna(df['social score'].mean())
```

```python
df =df.drop_duplicates()
```

```python
df['gender'].str.lower().str.strip()
```

```python
df.columns = df.columns.str.strip()
```

```python
df.describe()
```

```python
df['total'] = df['maths score'] + df['social score'] + df['science score']

df['average'] = df['total'] / 2
```

```python
topper_index = df['total'].idxmax()# topper ka index nikalana

# topper ka name aur score
topper_name = df.loc[topper_index, 'Nmae']
topper_score = df.loc[topper_index, 'total']

print(f"topper: {topper_name} with Toatl score: {topper_score}")
```

```python
colors = ['purple' if i !=topper_index else 'gold' for i in range (len(df))]

plt.bar(df['Nmae'], df['total'],color = colors)
plt.title("student total score")
plt.xlabel("students")
plt.ylabel("total score")
plt.show()
```

performance rating


```python
def performance_rating(avg):
  if avg >=90:
    return"excellent"
  elif avg >=75:
    return"good"
  elif avg >=60:
    return"average"
  else:
    return"poor"

df['performance_rating'] = df['average'].apply(performance_rating)
print(df)
```

```python
for col in ['maths score', 'science score', 'social score']:
  plt.plot(df['Nmae'], df[col], marker='o', label = col)
  plt.title("subject-wise Scores")
  plt.legend()
  plt.show()
```
