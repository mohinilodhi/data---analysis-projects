```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("employees.csv")
print(df.head())
```

```python

df["JoiningDate"] = pd.to_datetime(df["JoiningDate"], errors='coerce')
```

```python
print(df.dtypes)
```

```python
print("total employees:", len(df))
```

```python
print(df[df["Salary"] == df["Salary"].max()])
```

```python
print(df.groupby("Department")["Salary"].mean())
```

```python
dept_salary = df.groupby("Department")["Salary"].sum()
print(dept_salary)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df["Salary"], bins=5)
plt.title("Salary Distribution")
plt.show()
```

```python
sns.histplot(df["Department"], bins=5)
plt.title("Salary Distribution")
plt.show()
```

```python
sns.barplot(x="Department", y="Salary", data=df)
plt.xticks(rotation=45)
plt.title("Department vs Salary")
plt.show()
```

```python
sns.countplot(x="JoiningDate", data=df)
plt.xticks(rotation=45)
plt.title("Job Roles Count")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

plt.bar(df["Name"], df["Salary"])

plt.xlabel("Employee Name")
plt.ylabel("Salary")
plt.title("Name vs Salary")

plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(7,7))

plt.pie(df["Salary"], autopct="%1.1f%%")

plt.title("Salary Distribution")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

dept_salary.plot(kind="bar")

plt.title("Department vs Total Salary")
plt.xlabel("Department")
plt.ylabel("Total Salary")

plt.show()
```

```python
plt.figure(figsize=(10,5))
avg_salary = df.groupby("Department")["Salary"].mean()
avg_salary.plot(kind="bar")

plt.title("Average Salary per Department")
plt.xlabel("Department")
plt.ylabel("Average Salary")

plt.show()
```
