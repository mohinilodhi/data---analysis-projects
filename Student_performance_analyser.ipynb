```python
import pandas as pd

data = {
    "Name": ["A","B","C","D","E","F","G","H","I","J"],
    "Marks": [85,90,78,92,88,76,95,89,84,91],
    "Gender": ["Male","Female","Male","Female","Female","Male","Male","Female","Male","Female"],
    "StudyHours": [2,4,3,5,4,2,6,5,3,4]
}

df = pd.DataFrame(data)
```

```python
female_students = df[(df["Gender"] == "Female") & (df["Marks"] > 90)]
print(female_students)
```

```python
female_students_numeric_mean = df[(df["Gender"] == "Female")][["Marks", "StudyHours"]].mean()
print(female_students_numeric_mean)
```

```python
male_students_numeric_mean = df[(df["Gender"] == "Male")][["Marks", "StudyHours"]].mean()
print(male_students_numeric_mean)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.bar(df["Name"], df["Marks"])
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Student Marks")
plt.show()
```

```python
plt.scatter(df["StudyHours"], df["Marks"])
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Study vs Marks")
plt.show()
```

```python
gender_count = df["Gender"].value_counts()

plt.pie(gender_count.values, labels=gender_count.index, autopct="%1.1f%%")
plt.title("Gender Distribution")
plt.show()
```

```python
df["Performance"] = df["Marks"].apply(lambda x: "Good" if x > 85 else "Average")
print(df)
```
