```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('people-100.csv')
print(df)
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
df = df.dropna()
```

```python
df = df.drop_duplicates()
```

```python
df['First Name'] = df['First Name'].str.strip()
```

GRAPH OF DATA


```python
df['Sex'].value_counts().plot(kind='bar')
```

```python
df['domain'] = df["Email"].str.split('@').str[1]
df['domain'].value_counts().plot(kind='bar')
plt.title("Email Domians")
plt.show()
```

```python
df['domain'].value_counts().plot(kind='bar')
plt.title("Email Domains")
plt.show()
```

```python
df['domain'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    colors=['red','green','pink']
)
plt.title('Email Domain Distribution')
plt.show()
```
