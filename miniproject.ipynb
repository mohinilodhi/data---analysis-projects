```python
# Step 1: Import libraries
import matplotlib.pyplot as plt
import pandas as pd

# Step 2: Create sample dataset
data = {
    "Student": ["Aman", "Riya", "Karan", "Neha", "Vikas"],
    "Maths": [85, 90, 78, 92, 88],
    "Science": [80, 95, 70, 85, 89]
}

df = pd.DataFrame(data)

# Step 3: Print dataset
print("Student Marks Dataset:")
print(df)

# Step 4: Plot bar chart
plt.figure(figsize=(8,6))
plt.bar(df["Student"], df["Maths"], label="Maths", alpha=0.7)
plt.bar(df["Student"], df["Science"], label="Science", alpha=0.7)
plt.title("Student Marks Comparison")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.legend()
plt.show()
```
