```python
df['Total_Marks'] = df['Maths'] + df['Science'] + df['English'] + df['Sports']
print(df.Total_Marks)
```

```python
df['average'] = df['Total_Marks'] /20
print(df.average)
```

```python
import random
df['Grade'] = random.choice(['A', 'B', 'C', 'D', 'E'])
print(df.Grade)
```

```python
import random
df['acitvity marks'] =[random.randint(50, 100) for _ in range(20)]
print(df.head)
```

```python
topper = df[df['Total_Marks'] == df['Total_Marks'].max()]
print(topper)
```

```python
df['result'] = df['average'].apply(lambda x: 'fail' if x >= 35 else 'pass')
print(df.result)
```

```python
df = df.sort_values(by='Total_Marks', ascending=False)
print(df.head())
```
